Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
a=1
while a<=20:
  print(a)
  a=a+1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
n=int(input("enter no.."))
i=1
sum=0
while i<=n:
  sum=sum+i
  i=i+1
print(sum)
```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
n=int(input("enter any number..."))
rev=0
temp=n
while temp>0:
  digit=temp%10
  rev=rev*10+digit
  temp//=10
print("reversed number is ", rev)
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
num=int(input("enter number.."))
count=0
while num>0:
  num=num//10
  count+=1

print(count)
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
var= int(input())
new = var
res=0
while var>0:
  temp = var % 10
  res = res*10 + temp
  var = var // 10
print(res)
if new==res:
  print("Palindrome")
else:
  print("Not Palindrome")




```

Q8. Guess the number game.

Approach:
- Use while loop
- Continue until correct guess

```python
secret_number=123321
var=int(input("guess the number..."))
while secret_number!=var:
  print(" wrong gues...")
  var=int(input("try again..."))

print("right guess")
```
