## 2. Combine Two Numbers
Create two variables `a = 7` and `b = 3` and compute sum, difference, product, and a^b.

```python
a=7
b=3
sum=a+b
diff=a-b
pro=a*b
power=a**b
print("sum =",sum)
print("difference =",diff)
print("product=",pro)
print("power=",power)
```
